```python
import numpy as np
import pandas as pd

parking_demand = pd.read_csv("./data/서울특별시 광진구_자동차등록대수.csv",encoding="CP949")
df = pd.read_excel("./data/주차장(동별).xlsx",header=1, usecols=[2,4])
df
```

```python
df = df.rename(columns={"동별(3)":"읍면동명", "합계.1":"주차면수"})
df.drop([0,1,2],axis=0, inplace=True)
df.sort_values("읍면동명",inplace=True)
df.reset_index(inplace=True)
df.drop(columns=['index'],inplace=True)
```

```python
parking_demand["주차수요"] = parking_demand["자동차등록대수"]/df["주차면수"]
parking_demand
```

```python
parking_demand.to_excel("./data/주차수요.xlsx")
```
