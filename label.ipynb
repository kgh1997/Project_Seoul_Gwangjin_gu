### 군집 간의 점수 추출하기

```python
def get_score(series1, series2):
    '''
    두 시리즈를 인자로 받아서
    점수를 반환하는 함수입니다.
    '''
    result = series1*series2
    result.dropna(inplace=True)
    result = (result*100).sum()
    return result
```

```python
score = pd.DataFrame([get_score(df_0, parking_demand["주차수요"]),get_score(df_1,parking_demand["주차수요"]), get_score(df_2,parking_demand["주차수요"],), get_score(df_3,parking_demand["주차수요"],), get_score(df_4,parking_demand["주차수요"])], columns=["Score"], index=["0번군집","1번군집","2번군집","3번군집","4번군집"])
score
```

### 2번 군집 위치 찍어보기

```python
def get_map(df, file_name):
    '''
    Contains latitude and longitude data
    Receive a data frame df as an argument
    Get corresponding latitude and longitude in dataframe using folium
    This function returns a map and saves it as an HTML file.
    '''
    import folium
    map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)
    for index, row in df.iterrows():
        folium.Marker(location=[row['위도'], row['경도']]).add_to(map)
    map.save("./data/"+file_name + ".html")
    return map
```

```python
get_map(cluster2, "result")
```
