### 공영주차장과 가장 밀접한 군집 추출하기

```python
def get_closet_cluster(select, groupby_df, label, column_name):
    '''
    select: 공시지가_기준년월 제외 여부 변수
    공영주차장 데이터프레임과 
    그룹별로 집계화된 데이터프레임에서 각 클러스터링 별로
    가장 가까운 군집을 뽑아주는 함수입니다.
    '''
    from scipy.spatial import distance
    if select == 1:
        groupby_df = pd.concat([groupby_df.iloc[:,:8], groupby_df[column_name]], axis=1)
        # 각 군집의 평균값 계산
        groupby_df_means = groupby_df.groupby(column_name).mean()
        # 각 군집의 평균값과 공영주차장 데이터프레임의 평균값 사이의 거리 계산
        distances = distance.cdist(groupby_df_means, label.mean().values.reshape(1,-1))
        # 가장 가까운 군집 찾기
        closet_cluster = distances.argmin()
        print(f"가장 가까운 군집은 {closet_cluster}")
    if select == 2:
        groupby_df = pd.concat([groupby_df.iloc[:,:7], groupby_df[column_name]], axis=1)
        # 각 군집의 평균값 계산
        groupby_df_means = groupby_df.groupby(column_name).mean()
        # 각 군집의 평균값과 공영주차장 데이터프레임의 평균값 사이의 거리 계산
        distances = distance.cdist(groupby_df_means, label.drop("공시지가_기준년월", axis=1).mean().values.reshape(1,-1))
        # 가장 가까운 군집 찾기
        closet_cluster = distances.argmin()
        print(f"가장 가까운 군집은 {closet_cluster}")
    if select == 3:
        groupby_df = pd.concat([groupby_df.iloc[:,:6], groupby_df[column_name]], axis=1)
        # 각 군집의 평균값 계산
        groupby_df_means = groupby_df.groupby(column_name).mean()
        # 각 군집의 평균값과 공영주차장 데이터프레임의 평균값 사이의 거리 계산
        distances = distance.cdist(groupby_df_means, label.drop(columns=["공시지가_기준년월", "토지면적"]).mean().values.reshape(1,-1))
        # 가장 가까운 군집 찾기
        closet_cluster = distances.argmin()
        print(f"가장 가까운 군집은 {closet_cluster}")
    return groupby_df[groupby_df[column_name] == closet_cluster].mean()
```
