### 공영주차장과 가장 밀접한 군집 추출하기

```python
def get_closet_cluster(groupby_df, label, column_name):
    '''
    공영주차장 데이터프레임과 
    그룹별로 집계화된 데이터프레임에서 각 클러스터링 별로
    가장 가까운 군집을 뽑아주는 함수입니다.
    '''
    from scipy.spatial import distance
    # 각 군집의 평균값 계산
    groupby_df_means = groupby_df.groupby(column_name).mean()
    # 각 군집의 평균값과 공영주차장 데이터프레임의 평균값 사이의 거리 계산
    distances = distance.cdist(groupby_df_means, label.mean().values.reshape(1,-1))
    # 가장 가까운 군집 찾기
    closet_cluster = groupby_df_means.index[distances.argmin()]
    print(f"가장 가까운 군집은 {closet_cluster}")
    return groupby_df[groupby_df[column_name] == closet_cluster].mean()
```

```python
get_closet_cluster(adr_df.drop("Cluster2", axis=1), label, "Cluster1")
```
