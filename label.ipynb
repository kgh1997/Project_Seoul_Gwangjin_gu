### 공영주차장과 가장 밀접한 군집 추출하기

```python
def get_scaler(df):
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    scaled_df = scaler.fit_transform(df.values)
    scaled_df = pd.DataFrame(data=scaled_df, index=df.index, columns=df.columns)
    return scaled_df
```

#### 스케일링하기

```python
scaled_df = get_scaler(adr_df.drop(columns="Cluster9"))
scaled_label = get_scaler(label)
```

```python
# 스케일링된 데이터프레임에 군집라벨값 합치기
scaled_df["Cluster9"] = adr_df["Cluster9"]
```

```python
def get_distance(groupby_df, label, column_name):
    '''
    군집별로 정답 데이터와의 거리를 구하는 함수입니다.
    groupby_df: 군집 라벨 값이 포함된 데이터프레임
    label: 공영주차장의 데이터프레임
    column_name: 그룹으로 묶을 컬럼 이름인자
    '''
    import numpy as np
    import pandas as pd

    groupby_df = groupby_df.groupby(column_name).mean() # 각 군집별로 평균값을 가진 데이터프레임
    print(groupby_df)
    label_means = label.mean() # 정답 데이터의 평균
    print(label_means)
    distances = []
    for column in groupby_df.columns:
        distance =  np.sqrt(np.square(groupby_df.groupby(column_name).mean()[column] - label.mean()[column]))
        print(distance)
        distances.append(distance)
    distances = pd.DataFrame(distances).sum()
    return distances 
```

```python
def get_map(df):
    '''
    위도와 경도 데이터가 들어있는 
    데이터프레임 df를 인자로 받아서
    folium을 사용하여 데이터프레임의 해당 위경도를
    지도로 리턴하는 함수입니다.
    '''
    import folium
    map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)
    for index, row in df.iterrows():
        folium.Marker(location=[row['위도'], row['경도']]).add_to(map)
    return map
```
